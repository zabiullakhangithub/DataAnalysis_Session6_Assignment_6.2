```python
import pandas as pd
import matplotlib.pyplot as plt

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
labels = 'Male','Female'

# Get male and female counts from sex column
sizes = titanic.sex.value_counts()
#print (sizes)

#Create a pie chart presenting the male/female proportion
colors=['purple','red']
plt.axis('equal')
plt.title('Male/Female Proportion')
plt.pie(sizes, explode=(0.01,0), labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()

#Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(9,6))
ax = plt.subplot()
ax.scatter(titanic[titanic['sex']=='male']['age'],titanic[titanic['sex']=='male']['fare'],c='purple',s=40)
ax.scatter(titanic[titanic['sex']=='female']['age'],titanic[titanic['sex']=='female']['fare'],c='red',s=40)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.legend(('male','female'),scatterpoints=1,loc='upper right',fontsize=15,)

```
